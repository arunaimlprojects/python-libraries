Pandas **Series** एक **one-dimensional array** होती है, जो किसी भी डेटा टाइप (integer, float, string, object, etc.) को स्टोर कर सकती है। इसमें **NumPy array** जैसी indexing और slicing का सपोर्ट होता है।  

### **Fundamental Steps in Pandas Series**  

#### **1. Import Pandas**  
```python
import pandas as pd
```

#### **2. Create a Pandas Series**  
```python
data = pd.Series([10, 20, 30, 40, 50])  # Numeric Series
print(data)
```
##### **Output:**
```
0    10
1    20
2    30
3    40
4    50
dtype: int64
```

---

#### **3. Custom Indexing in Series**  
```python
data = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(data)
```
##### **Output:**
```
a    10
b    20
c    30
dtype: int64
```

---

#### **4. Accessing Elements (Indexing & Slicing)**  
```python
print(data[0])   # Access first element
print(data["b"]) # Access element using custom index
print(data[:2])  # Slice first two elements
```

---

#### **5. Mathematical Operations**  
```python
print(data + 10)   # Add 10 to each element
print(data * 2)    # Multiply each element by 2
```

---

#### **6. Checking for Null Values**  
```python
data = pd.Series([10, 20, None, 40, 50])
print(data.isnull())  # Check for missing values
```
##### **Output:**
```
0    False
1    False
2     True
3    False
4    False
dtype: bool
```

---

#### **7. Converting Series to List**  
```python
data_list = data.tolist()
print(data_list)  # Output: [10, 20, None, 40, 50]
```

---

#### **8. Filtering Values in a Series**  
```python
filtered_data = data[data > 20]  # Select elements greater than 20
print(filtered_data)
```

---

#### **9. Apply Function to Series**  
```python
print(data.apply(lambda x: x * 2))  # Multiply each element by 2
```

---

#### **10. Get Basic Statistics**  
```python
print(data.describe())  # Get count, mean, std, min, max, etc.
```

---
